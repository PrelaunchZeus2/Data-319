#### 3. Suppose there are 100 items, numbered 1 to 100, and also 100 baskets, also numbered 1 to 100. Item i is in basket b if and only if i divides b with no remainder. Thus, item 1 is in all the baskets, item 2 is in all fifty of the even-numbered baskets, and so on. Basket 12 consists of items $\{1, 2, 3, 4, 6, 12\}$, since these are all the integers that divide 12. Answer the following questions:

##### (a) If the basket is a prime number, how many items will it consist of?

##### (b) Which basket is the largest (i.e. it consists of the largest count of items)?

##### (c) If the support threshold is 6, which items are frequent?

##### (d) Apply the A-Priori Algorithm with support threshold 6 to this data to determine all frequent itemsets.

See Code Below

##### (e) What is the interest of the following association rules?
i. $\{5, 7\}\rightarrow2$ = 0.5

ii. $\{2, 3, 4\}\rightarrow5$ = 0.125

*Note that you are encouraged but not required to use Python to obtain the answers for these questions. If you opt out from using Python in favor of "paper-and-pencil" (i.e. no coding involved) approach, then show your work to get full credit (reporting just answers will not be sufficient)*

```python
#question 3
from collections import Counter
baskets = [[i for i in range(1,101) if b % i == 0] for b in range(1,101)]
flat_baskets = [item for basket in baskets for item in basket]
counts = Counter(flat_baskets)

#A
primes = []
for n in range(1, 101):
    if n > 1:
        for i in range(2, int(n**0.5) + 1):
            if (n % i) == 0:
                break
        else: 
            primes.append(n)

prime_basket_counts = {prime: len(baskets[prime-1]) for prime in primes}

for prime, count in prime_basket_counts.items():
    print(f"Prime Number {prime} Basket has {count} items.")
    
#B
Max = 0
largest_basket = 0

for n in range(1, 101):
    divisors = [i for i in range(1, n + 1)]
    if len(divisors) > Max:
        Max = len(divisors)
        largest_basket = n
    
print(f"Largest Basket: {largest_basket}")

#C
frequent_itemsets = [item for item, count in counts.items() if count > 6]
print(f"\nFrequent Itemsets: {frequent_itemsets}")

#D
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
import pandas as pd
pd.set_option('display.max_rows', None)
te = TransactionEncoder()
te_ary = te.fit(baskets).transform(baskets)
df = pd.DataFrame(te_ary, columns=te.columns_)

priori_sets = apriori(df, min_support = 0.06, use_colnames = True)
priori_sets['length'] = priori_sets['itemsets'].apply(lambda x: len(x))
priori_sets = priori_sets[priori_sets['length'] >= 6].reset_index(drop = False)
print(f"\nApriori Frequent Sets: {priori_sets}")

#E
def cal_sup(itemset, baskets):
    count = sum(1 for basket in baskets if all(item in basket for item in itemset))
    return count / len(baskets)

#i
sup_5_7 = cal_sup({5, 7}, baskets)
sup_2 = cal_sup({2}, baskets)
sup_5_7_2 = cal_sup({5, 7, 2}, baskets)
interest = sup_5_7_2 / sup_5_7
print(f"\nInterest 5, 7 -> 2: {interest}")

 
#ii
sup_2_3_4 = cal_sup({2, 3, 4}, baskets)
sup_5 = cal_sup({5}, baskets)
sup_2_3_4_5 = cal_sup({2, 3, 4, 5}, baskets)
interest = sup_2_3_4_5 / sup_2_3_4
print(f"Interest 2, 3, 4 -> 5: {interest}")

```
